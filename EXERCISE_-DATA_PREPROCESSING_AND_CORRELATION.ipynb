1: IMPUTING NULL VALUES
    
    

```python

# Impute missing values with the mode (most frequent value) in column 'color':
data['color']=data['color'].fillna(data['color'].mode().iloc[0])

data.describe()
```

```python

#Impute missing values with mean in coloumn 'carat':
data['carat']=data['carat'].fillna(data['carat'].mean())

data.info()
```

```python
#imputing missing values with meadian in column 'table':

data['table']=data['table'].fillna(data['table'].median())

data.info()
```

```python
data.isna().sum()
```

2 ENCODING CATEGORICAL COLUMNS 

```python
from sklearn.preprocessing import LabelEncoder
import pandas as pd

column_to_encode = 'cut'
column_to_encode_values = data[column_to_encode]

# Initialize the LabelEncoder
label_encoder = LabelEncoder()

# Fit and transform the column values to obtain encoded labels
encoded_labels = label_encoder.fit_transform(column_to_encode_values)

# Add a new column for the encoded labels
data['encoded_cut'] = encoded_labels

# Display the resulting DataFrame
print(data)
```

```python
column_to_encode = 'color'
column_to_encode_values = data[column_to_encode]

label_encoder = LabelEncoder()

encoded_labels = label_encoder.fit_transform(column_to_encode_values)

data['encoded_color'] = encoded_labels

print(data)
```

```python
column_to_encode = 'clarity'
column_to_encode_values = data[column_to_encode]

label_encoder = LabelEncoder()

encoded_labels = label_encoder.fit_transform(column_to_encode_values)

data['encoded_clarity'] = encoded_labels

print(data)
```

```python
data=data.drop(['cut','color','clarity'],axis=1)

data
```
